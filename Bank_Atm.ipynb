```python
all_months_data = pd.DataFrame()
all_months = pd.DataFrame()
```

```python
import numpy as np
for file in files:
    xl = pd.ExcelFile('D:\\bank_project\\'+file)
    all_months = pd.read_excel('D:\\bank_project\\'+file,header = [1,2,3] ,skiprows = [0,5] )
    all_months['date'] = pd.to_datetime(xl.sheet_names)[0]
    all_months = all_months[['Bank Name','ATMs','POS','Credit Cards','Debit Cards','date']]
    idx = all_months.loc[all_months['Bank Name', 'Unnamed: 2_level_1', 'Unnamed: 2_level_2']=='Total'].index.values[0]
    all_months_df = all_months[['Bank Name','ATMs','POS','Credit Cards','Debit Cards','date']][:idx]
    col_in = ('Credit Cards','Amount of transactions\n(Lakh)') in all_months_df.columns
    if col_in == True:
        d = {'Amount of transactions\n(Lakh)': 'Amount of transactions\n(Rs. Millions)'}
        all_months_df = all_months_df.rename(columns=d, level=1)
        all_months_df['Debit Cards','Amount of transactions\n(Rs. Millions)'] = all_months_df['Debit Cards','Amount of transactions\n(Rs. Millions)'].div(10)
        all_months_df['Credit Cards','Amount of transactions\n(Rs. Millions)'] = all_months_df['Credit Cards','Amount of transactions\n(Rs. Millions)'].div(10)
    d = {'No .of outstanding cards as at the end of the month': 'No. of outstanding cards as at the end of the month'}
    all_months_df = all_months_df.rename(columns=d, level=1)
    all_months_data = pd.concat([all_months_data,all_months_df])
```

```python
#processing 2020 march to available data
all_2020_data = pd.DataFrame()
all_months_2020_data = pd.DataFrame()
files = [file for file in os.listdir('D:\\2020')]

for idx, file in enumerate(files):
    xl = pd.ExcelFile('D:\\2020\\'+file)
    all_2020_data = pd.read_excel('D:\\2020\\'+file,header = [1,2,3] ,skiprows = [0,5,6,7,27,48,57,64]) 
    col = ('PoS') in all_2020_data.columns
    if col == True:
        d = {'PoS': 'POS'}
        all_2020_data = all_2020_data.rename(columns=d)
    all_2020_data = all_2020_data[['Bank Name','ATMs','POS','Credit Cards','Debit Cards']]
    all_2020_data['date'] = pd.to_datetime(xl.sheet_names)[0]
    #if idx==0 or idx==8:
    if ('Bank Name', 'Unnamed: 2_level_1', 'Unnamed: 2_level_2') in all_2020_data.columns:
        idx = all_2020_data.loc[all_2020_data['Bank Name', 'Unnamed: 2_level_1', 'Unnamed: 2_level_2']=='Total'].index.values[0]
    else:
        idx = all_2020_data.loc[all_2020_data['Bank Name', 'Unnamed: 1_level_1', 'Unnamed: 1_level_2']=='Total'].index.values[0]
    all_2020_data = all_2020_data[['Bank Name','ATMs','POS','Credit Cards','Debit Cards','date']][:idx]   
    if ('Credit Cards','Value of transactions\n(Rupees Lakh)') in all_2020_data.columns:
        d = {'Value of transactions\n(Rupees Lakh)': 'Amount of transactions\n(Rs. Millions)'}
        all_2020_data = all_2020_data.rename(columns=d,level=1)
    else:
        d = {'Amount of transactions\n(Lakh)': 'Amount of transactions\n(Rs. Millions)'}
        all_2020_data = all_2020_data.rename(columns=d,level=1)
    all_2020_data['Debit Cards','Amount of transactions\n(Rs. Millions)'] = all_2020_data['Debit Cards','Amount of transactions\n(Rs. Millions)'].div(10)
    all_2020_data['Credit Cards','Amount of transactions\n(Rs. Millions)'] = all_2020_data['Credit Cards','Amount of transactions\n(Rs. Millions)'].div(10)
    d = {'No .of outstanding cards as at the end of the month': 'No. of outstanding cards as at the end of the month'}
    all_2020_data = all_2020_data.rename(columns=d, level=1)
    if ('Debit Cards','No. of outstanding cards as at the end of the month','Unnamed: 13_level_2') in all_2020_data.columns:
        d = {'Unnamed: 13_level_2': 'Unnamed: 11_level_2'}
        all_2020_data = all_2020_data.rename(columns=d,level=2) 
    if ('Credit Cards','No. of outstanding cards as at the end of the month','Unnamed: 8_level_2') in all_2020_data.columns:
        d = {'Unnamed: 8_level_2': 'Unnamed: 6_level_2'}
        all_2020_data = all_2020_data.rename(columns=d,level=2) 
    all_months_2020_data = pd.concat([all_months_2020_data,all_2020_data])
    
```
